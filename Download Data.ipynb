```python
import arrow
import requests 

stocks = pd.read_csv('bovespa.csv')

data_range ='5y'
time_interval='1d'
folder = 'data/'

for stock in stocks['codigo.sa'].values:
    
    
    res = requests.get('https://query1.finance.yahoo.com/v7/finance/chart/' + stock + '?&range=' + data_range + 
                       '&interval=' + time_interval.format(**locals()))
    data = res.json()

    body = data['chart']['result'][0]
    dt = datetime.datetime
    dt = pd.Series(map(lambda x: arrow.get(x).to('EST').datetime.replace(tzinfo=None), body['timestamp']), name='dt')
    df = pd.DataFrame(body['indicators']['quote'][0], index=dt)


    df['data'] = df.index
    df.reset_index(drop=True)

#     df.to_csv(folder + stock + '.csv')
```
