```python
import sys

import numpy as np
import pandas as pd
import functions as DLmodels          

# dados_polarity = pd.read_csv('../logs/data-polarity.csv')
dados_mape_self = pd.read_csv('../logs/data_MAPE_self.csv')
dados_mape_self['Type'] = 'dados_mape_self'

dados_mape_self_complete = pd.read_csv('../logs/data_MAPE_self_complete.csv')
dados_mape_self_complete['Type'] = 'dados_mape_self_complete'

dados_mape_mix = pd.read_csv('../logs/data_MAPE_mix.csv')
dados_mape_mix['Type'] = 'dados_mape_mix'

dados_future_self = pd.read_csv('../logs/data_future_self.csv')
dados_future_self['Type'] = 'dados_future_self'
dados_future_self['MAPE'] = dados_future_self['gain_pred']


dados_future_self_complete = pd.read_csv('../logs/data_future_self_complete.csv')
dados_future_self_complete['Type'] = 'dados_future_self_complete'
dados_future_self_complete['MAPE'] = dados_future_self_complete['gain_pred']

dados_future_mix = pd.read_csv('../logs/data_future_mix.csv')
dados_future_mix['Type'] = 'dados_future_mix'
dados_future_mix['MAPE'] = dados_future_self['gain_pred']

```

```python


dados_mape_self = dados_mape_self[['Stock',	'ML',	'MAPE', 'Type']]      
dados_mape_mix = dados_mape_mix[['Stock',	'ML',	'MAPE', 'Type']]      
dados_mape_self_complete= dados_mape_self_complete[['Stock',	'ML',	'MAPE', 'Type']]      



data_mape = pd.concat([dados_mape_self,
           dados_mape_mix,
           dados_mape_self_complete
          ])
data_mape = data_mape[(data_mape['MAPE'] < 5)]


dados_future_self = dados_future_self[['Stock',	'ML',	'MAPE', 'Type']]      
dados_future_mix = dados_future_mix[['Stock',	'ML',	'MAPE', 'Type']]      
dados_future_self_complete= dados_future_self_complete[['Stock',	'ML',	'MAPE', 'Type']]      



data_future = pd.concat([
           dados_future_self,
           dados_future_mix,
           dados_future_self_complete
          ])


data_mape['ML'].unique()
data = pd.concat([data_mape, data_future[data_future['Stock'].isin(data_mape['ML'].unique())]])
data
```
